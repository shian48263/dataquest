```python
recent_grads['ShareMen'] = recent_grads['Men'] / recent_grads['Total']
```

```python
arts = recent_grads[recent_grads["Major_category"] == "Arts"]
arts.set_index("Major", inplace=True)
arts[["ShareMen", "ShareWomen"]].plot(kind="bar")
plt.show()
```

```python
import numpy as np
arts = recent_grads[recent_grads["Major_category"] == "Arts"]
```

```python
fig_bar = plt.figure(figsize=(8,8))
bar = fig_bar.add_subplot(1,1,1)
locs = np.arange(0, len(arts['Major']))
bar_1 = bar.bar(left=locs, height=arts['ShareMen'].tolist(), width=0.35)
bar.set_xticklabels(arts['Major'].values, rotation=90)
plt.show()
```

```python
fig_bar = plt.figure(figsize=(8,8))
bar = fig_bar.add_subplot(1,1,1)
locs = np.arange(0, len(arts['Major']))
bar_1 = bar.bar(left=locs, height=arts['ShareMen'].tolist(), width=0.35)
offset_locs = locs + 0.35
bar_2 = bar.bar(left=offset_locs, height=arts['ShareWomen'].tolist(), width=0.35, color='green')
bar.set_xticks(offset_locs)
bar.set_xticklabels(arts['Major'].values, rotation=90)
plt.legend((bar_1, bar_2), ("ShareMen", "ShareWomen"), loc="upper left")
plt.grid()
plt.show()
```
